```python
%sql
WITH cte1 AS (
  SELECT
    `invc_nbr`,
    `mat_nbr`,
    cal_month,
    cal_month,
    COUNT(*) AS duplicate_count
  FROM
    adh_genpro_use2_prd.g_order360.billing_transactions_hhd
  WHERE
    zzrefinv != ''
    AND bill_type IN ('zrk1', 'zrk2')
    AND ZPRICEOWN IN ('10', '20', '30', '40')
    AND salesorg IN ('1000', '2000', '1200')
    AND CAST(CREATED AS DATE) BETWEEN
      DATEADD(day, -180, current_timestamp())
    AND
      current_timestamp()
  GROUP BY
    `invc_nbr`,
    `mat_nbr`,
    cal_month,
    cal_month
)
SELECT
  'BILL_NUM || BILL_ITEM || FISCVARNT || CALMONTH' AS PRIMARY_KEY_NAME,
  CAST(invc_nbr AS STRING)
    || ' || '
    || CAST(mat_nbr AS STRING)
    || ' || '
    || CAST(cal_month AS STRING)
    || ' || '
    || CAST(cal_month AS STRING) AS PRIMARY_KEY_VALUE,
  'BILL_NUM || BILL_ITEM || FISCVARNT || CALMONTH' AS ATTRIBUTE_NAME,
  CAST(invc_nbr AS STRING)
    || ' || '
    || CAST(mat_nbr AS STRING)
    || ' || '
    || CAST(cal_month AS STRING)
    || ' || '
    || CAST(cal_month AS STRING) AS ATTRIBUTE_VALUE,
  'duplicate_count' AS ADDITIONAL_COLUMN_NAME,
  duplicate_count AS ADDITIONAL_COLUMN_VALUE
FROM
  cte1
```
